# **ðŸ§  ReduÃ§Ã£o de Dimensionalidade em Imagens para Redes Neurais**

Este projeto demonstra como reduzir a complexidade de imagens para uso em redes neurais. A imagem colorida Ã© convertida para escala de cinza e, em seguida, binarizada (preto e branco), simplificando os dados visuais.

Essas transformaÃ§Ãµes facilitam o processamento pelas redes neurais, pois eliminam informaÃ§Ãµes desnecessÃ¡rias e reduzem a dimensionalidade da entrada, tornando o treinamento mais eficiente.



**âœ… Passo 1 â€“ Importar as bibliotecas no Python**

Essas bibliotecas permitem carregar, processar e exibir imagens. cv2 Ã© do OpenCV e matplotlib.pyplot serve para mostrar imagens no Jupyter ou Colab.

```python
import cv2
import matplotlib.pyplot as plt
```

âœ… **Passo 2 â€“ Carregar a imagem**

Aqui vocÃª carrega uma imagem colorida do seu computador. O OpenCV vai ler no formato BGR (azul, verde, vermelho).

```python
# LÃª a imagem colorida
imagem_colorida = cv2.imread("imagem.png")  # substitua pelo nome da sua imagem
```

âœ… **Passo 3 â€“ Converter para escala de cinza**

Essa conversÃ£o transforma a imagem colorida em tons de cinza (de 0 a 255), usando uma mÃ©dia ponderada das cores RGB.

```python
# Converte para escala de cinza
imagem_cinza = cv2.cvtColor(imagem_colorida, cv2.COLOR_BGR2GRAY)
```

âœ… **Passo 4 â€“ Aplicar binarizaÃ§Ã£o (Preto e Branco)**

A imagem cinza Ã© transformada em preto e branco com base em um valor de corte (limiar). Pixels acima do limiar viram branco (255), os demais viram preto (0).

```python
# Define o limiar (threshold) para binarizaÃ§Ã£o
limiar = 127

# Aplica binarizaÃ§Ã£o: pixels acima do limiar viram 255, abaixo viram 0
_, imagem_binarizada = cv2.threshold(imagem_cinza, limiar, 255, cv2.THRESH_BINARY)
```

âœ… **Passo 5 â€“ Exibir os resultados**

Esse trecho exibe as trÃªs versÃµes da imagem: original, em tons de cinza e binarizada. Convertemos de BGR para RGB sÃ³ para mostrar corretamente no Matplotlib.

```python
# Converte imagem colorida para RGB sÃ³ para exibiÃ§Ã£o correta com matplotlib
imagem_rgb = cv2.cvtColor(imagem_colorida, cv2.COLOR_BGR2RGB)

# Exibe as trÃªs imagens lado a lado
plt.figure(figsize=(15,5))

plt.subplot(1, 3, 1)
plt.title("Imagem Colorida")
plt.imshow(imagem_rgb)
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title("Escala de Cinza")
plt.imshow(imagem_cinza, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title("Imagem Binarizada")
plt.imshow(imagem_binarizada, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
